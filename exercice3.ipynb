```python
import pandas as pd
import matplotlib.pyplot as plt

# Données fictives : ventes d'une boutique
data = {
    "mois":    ["Jan","Fév","Mar","Avr","Mai","Jun","Jul","Aoû","Sep","Oct","Nov","Déc"],
    "ventes":  [320, 290, 410, 380, 450, 520, 610, 580, 430, 370, 490, 680],
    "budget":  [350, 300, 400, 400, 450, 500, 600, 550, 400, 380, 500, 700],
    "region":  ["Nord","Nord","Sud","Sud","Est","Est","Ouest","Ouest","Nord","Sud","Est","Ouest"]
}
df = pd.DataFrame(data)
print(df)


```

```python
df.head(3)          # les 3 premières lignes
df.tail(3)          # les 3 dernières lignes
df.shape            # dimensions du tableau
df.dtypes           # type de chaque colonne
df.describe()       # statistiques automatiques
df["ventes"].mean() # moyenne des ventes
df["ventes"].max()  # vente maximale

# Filtrer : mois où les ventes > 500
df[df["ventes"] > 500]

# Nouveau calcul : écart budget vs réel
df["ecart"] = df["ventes"] - df["budget"]
print(df[["mois", "ventes", "budget", "ecart"]])

```

```python
plt.figure(figsize=(10, 4))
plt.plot(df["mois"], df["ventes"], marker="o", color="steelblue", label="Ventes réelles")
plt.plot(df["mois"], df["budget"], marker="s", color="orange", linestyle="--", label="Budget")
plt.title("Ventes vs Budget — Boutique fictive")
plt.xlabel("Mois")
plt.ylabel("Unités vendues")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```
