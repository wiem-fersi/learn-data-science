```python
# Les listes (comme un tableau à une dimension)
ventes = [150, 230, 180, 410, 95]
print(ventes[0])          # → 150 (les indices commencent à 0 !)
print(len(ventes))        # → 5 (nombre d'éléments)

```

```python

# Les conditions
if age >= 18:
    print("Majeur")
else:
    print("Mineur")

```

```python
# Les boucles
for vente in ventes:
    print(vente * 1.2)    # affiche chaque vente + 20%
```

```python
# Les fonctions
def calculer_moyenne(liste):
    return sum(liste) / len(liste)

print(calculer_moyenne(ventes))  # → 213.0
```

```python
# Créer un DataFrame (tableau de données)
data = {
    "produit": ["Chaussures", "Sac", "Ceinture", "Chapeau"],
    "ventes":  [150, 89, 230, 45],
    "prix":    [79.99, 49.99, 29.99, 19.99]
}
df = pd.DataFrame(data)

# Explorer les données
print(df.head())          # affiche les 5 premières lignes
print(df.shape)           # → (4, 3) : 4 lignes, 3 colonnes
print(df.info())          # types de données et valeurs manquantes
print(df.describe())      # statistiques descriptives automatiques

# Filtrer et trier
df_chers = df[df["prix"] > 40]           # produits > 40€
df_tries = df.sort_values("ventes", ascending=False)

# Calculer une nouvelle colonne
df["chiffre_affaires"] = df["ventes"] * df["prix"]
print(df["chiffre_affaires"].sum())       # CA total

```
